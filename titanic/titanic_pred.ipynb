```python
def load_data(file):
    
    df = pd.read_csv(file)
    df = df.drop('Cabin', axis=1)
    df = df.drop('Ticket', axis=1)
    df = df.drop('Name', axis=1)

    for i in range(df.shape[0]):
        if df.iloc[i]['Sex'].strip() == 'male':
            df.at[i,'Sex'] = 0
        else:
            df.at[i,'Sex'] = 1

        if df.iloc[i]['Embarked'] == 'S':
            df.at[i, 'Embarked'] = 0
        elif df.iloc[i]['Embarked'] == 'C':
            df.at[i, 'Embarked'] = 1
        else:
            df.at[i, 'Embarked'] = 2
    df = df.dropna()
    
    data = df.to_numpy()
    labels = data[:, 1]
    labels = labels.reshape(labels.shape[0],1)
    data = np.delete(data, 1, 1)
    
    data = data.astype('double')
    labels = labels.astype('double')    
    return data, labels
```

```python
model = Net(input_channel=1)
train_model(model, train, train_ys)
```
